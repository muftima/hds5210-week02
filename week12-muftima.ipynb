Use the Python mysqlclient module as demonstrated in class to query the cdi table to describe how many unique (distinct) values there are for each of the fields (columns) below:
DataSource

```python
import pymysql.cursors

connection = pymysql.connect(
    host='localhost',
    user='slucor',
    password='S1UC0R',
    db='hds5210',
    cursorclass=pymysql.cursors.DictCursor)

cursor = connection.cursor()
# Read a single record
sql = "SELECT Datasource, DataValueType ,LocationAbbr ,Question, Topic ,TopicID, YearEnd FROM cdi"
cursor.execute(sql)
result = cursor.fetchall()
connection.close()
```

```python
result
```

```python
distinct = {}

for row in result:
    for key, value in row.items():
        if key not in distinct:
            distinct[key] = []
        if value not in distinct[key]:
            distinct[key].append(value)

for name, columns in distinct.items():
    print(name)
    print(len(columns))
```
